# EDA of Text Data

With the text data that we have in front of us, one of the things we can do is to check what words are frequently used in the real news headlines, and compare that against words that are frequently used in fake news headlines. To do this, we can create Word Clouds for the fake and real news headlines.

Furthermore, while reading through the data, one could see that the fake news had more negative connotations, so we used 



```python
#Testing Import of WordCloud - It Works! Poggers
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from textwrap import wrap
from sklearn.feature_extraction.text import CountVectorizer
real_data = pd.read_csv('cleaned_True.csv')
fake_data = pd.read_csv('cleaned_Fake.csv')

wordcloud2 = WordCloud().generate(''.join(real_data['title']))
f = plt.figure(figsize=(10,10))
plt.imshow(wordcloud2)
plt.axis("off")
plt.title("True Data WordCloud")
plt.show()

wordcloud3 = WordCloud().generate(''.join(fake_data['title']))
plt.imshow(wordcloud3)
plt.axis("off")
plt.title("Fake Data WordCloud")
plt.show()
```

## Observations from the WordCloud
1. Both Datasets have lot of influence from politics
2. True data also refers to happenings in other countries and the WordCloud seems to have a more narrative tone. We can see that say is one of the most common words in the true data WordCloud, hinting that it has more reporting style, while Fake News WordCloud does not have this, hinting that it does not have the same evidences that are present in Real News.
3. Fake News focuses more on the United States and specifially about politicians like Barrack Obama, Donald Trump, Hilary Clinton and political parties - Democrats and Republicans.
